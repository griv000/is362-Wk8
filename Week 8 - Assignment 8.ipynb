```python
# import modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pyplot
%matplotlib inline
```

```python
# Create list with headers as they are not included in the csv file
mpg_atributes = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin', 'car name']
```

```python
# import data csv, split columns and use list for header
mpg = pd.read_csv("auto-mpg.data", delim_whitespace=True, header=None, names=mpg_atributes)
mpg
```

```python
# replace '?' in the horsepower column with NaN and convert to numeric
mpg.horsepower = mpg.horsepower.replace('?', np.NaN)
mpg.horsepower = pd.to_numeric(mpg.horsepower)
```

```python
# Replace 1, 2, 3 in origin with names
mpg.origin.replace([1,2,3], ['USA','Asia','Europe'], inplace=True)
mpg
```

```python
# Create bar chart showing distribution of cylinders
mpg.hist(column='cylinders')
pyplot.suptitle('Distribution of')
pyplot.xlabel('# of Cylinders')
pyplot.ylabel('# Cars')
pyplot.show()
```

```python
# Create a scatterplot that shows the relationship between horsepower and weight
pyplot.scatter(mpg.horsepower, mpg.weight)
pyplot.xlabel('weight')
pyplot.ylabel('horsepower')
pyplot.show()
```

```python
# Custom question using seaborn: what is the relationship between horsepower and cylinders, broken down by country?
sns.lmplot('cylinders', 'horsepower', data=mpg, hue='origin')
pyplot.show()
```
